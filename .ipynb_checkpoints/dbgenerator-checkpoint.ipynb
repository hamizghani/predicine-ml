```python
medicine_names = [
    "Paracetamol", "Ibuprofen", "Aspirin", "Naproxen", "Ketoprofen", "Amoxicillin", "Azithromycin", "Ciprofloxacin",
    "Doxycycline", "Clindamycin", "Oseltamivir", "Acyclovir", "Remdesivir", "Favipiravir", "Zanamivir", "Metformin",
    "Insulin", "Glipizide", "Sitagliptin", "Pioglitazone", "Atorvastatin", "Simvastatin", "Lisinopril", "Amlodipine",
    "Losartan", "Furosemide", "Hydrochlorothiazide", "Spironolactone", "Warfarin", "Clopidogrel", "Dexamethasone",
    "Prednisone", "Hydrocortisone", "Methotrexate", "Mycophenolate", "Levothyroxine", "Propylthiouracil", "Methimazole",
    "Fluoxetine", "Sertraline", "Diazepam", "Lorazepam", "Alprazolam", "Haloperidol", "Risperidone", "Omeprazole",
    "Ranitidine", "Pantoprazole", "Esomeprazole", "Lansoprazole"
]

ids = [f"D{i:03}" for i in range(1, len(medicine_names) + 1)]
briefs = [f"{name} is a commonly used medication." for name in medicine_names]
descriptions = [
    f"{name} is used for treating various conditions. It works through specific mechanisms depending on its category. It is widely prescribed and generally well tolerated."
    for name in medicine_names
]

df = pd.DataFrame({
    "id": ids,
    "medicine_name": medicine_names,
    "brief": briefs,
    "description": descriptions,
    "photo_link": [None] * len(medicine_names)
})

# Tampilkan hasil
print(df.head())
```
